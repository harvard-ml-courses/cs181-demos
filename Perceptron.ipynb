```python
import matplotlib.pyplot as plt
import numpy.random
import numpy as np
import seaborn
%pylab inline
pylab.rcParams['figure.figsize'] = (8, 8)
#seaborn.set_context("talk")
```

```python
n   = 50
X   =  (numpy.random.random([n, 2])) * 4
tW  = np.array([0.5, 1]).T
tW0 = -3
# Classification of the points
y   = np.sign(np.dot(X, tW) + tW0)
```

```python
def graph(X, y, w, w0, oldw=None, oldw0=None, x=None, up=None):
    plt.figure()
    gx = np.linspace(0, 4, 100).transpose()
    plt.plot(gx, -(gx * w[0] + w0)/ w[1])
    plt.scatter(X[y < 0][:, 0], X[y < 0][:, 1], color="red")
    plt.scatter(X[y > 0][:, 0], X[y > 0][:, 1], color="blue")


    

    plt.xlim([-1, 5])
    plt.ylim([-1, 5])
    plt.quiver([0], [0], w[0], w[1], angles='xy',scale_units='xy',scale=1)

    if oldw is not None:
        plt.plot(gx, -(gx * oldw[0] + oldw0)/ oldw[1], color="green", linestyle="dotted")        
        plt.quiver(oldw[0], oldw[1], up[0], up[1], angles='xy',scale_units='xy',scale=1,linestyle ="dotted")
        plt.quiver(0, 0, oldw[0], oldw[1], angles='xy',scale_units='xy',scale=1,linestyle ="dotted")
        plt.scatter(x[ 0], x[ 1], color="black", marker="o")
```

```python
graph(X, y, tW, tW0)
plt.title("True Separator")
```

```python
import sklearn.linear_model
model = sklearn.linear_model.LinearRegression().fit(X, y)
graph(X, y, model.coef_, model.intercept_)
plt.title("Least Squares Separator")
```

```python
X2 = np.vstack([X, [3,30]])
y2 = np.hstack([y, 1])
model = sklearn.linear_model.LinearRegression().fit(X2, y2) 
graph(X2, y2, model.coef_, model.intercept_)
plt.title("Least Squares Separator with Outlier")
```

```python
num_wrong = 1
ii = 0
W = np.array([0.5, 1])
W0 = 0
while num_wrong > 0:
  ii = ii + 1

  print('Epoch ', ii)

  num_wrong = 0


  for i in range(n):
    # classify point
    a = np.dot(X[i,:], W) + W0

    # old weights
    oldW = np.copy(W)
    oldW0 = np.copy(W0)
    
    # Classified incorrectly
    if a*y[i] <= 0:
      W = W + y[i] * X[i, :].T
      W0 = W0 + y[i]
      print('\tDatum %02d/%02d:\ta=% 4.4f\tt=% 2d [incorrect]'%( i, n, a, y[i]) )
      print('\t\tAdjust w = w + [% 4.4f % 4.4f]'%(y[i]*X[i,0], y[i]*X[i,1]))
      print('\t\tAdjust w0 = w0 + % 4.4f'%( y[i]))
      num_wrong = num_wrong + 1;
      graph(X, y, W, W0, oldW, oldW0, X[i,:], y[i] * X[i, :].T)
      plt.title("Epoch %d Datum %d"%(ii, i))
    else:
      print('\tDatum %02d/%02d:\ta=% 4.4f\ty=% 2d [correct]'%( i, n, a, y[i]) )
     
```
